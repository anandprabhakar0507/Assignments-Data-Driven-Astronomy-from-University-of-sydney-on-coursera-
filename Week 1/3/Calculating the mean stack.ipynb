```python
%matplotlib inline
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
import sys
```

```python
def mean_datasets(arrays):
    dataset = None
    for array in arrays:
        if dataset is None:
            dataset = array
        else:
            dataset += array
    meanset = dataset / len(arrays)

    return meanset
```

```python
def load_fits(path):
    with fits.open(path) as hdulist:
        my_data = hdulist[0].data
        return my_data
```

```python
def mean_fits(paths):
    fits_array = []
    for path in paths:
        fits_array.append(load_fits(path))
        
    return mean_datasets(fits_array)
```

```python
from os import listdir
from os.path import isfile, join
mypath = 'fits-img-mean'
fitsfiles = [join(mypath, f) for f in listdir(mypath) if isfile(join(mypath, f))]
fitsfiles
```

```python
data = mean_fits(fitsfiles[:500])
    
# You can also plot the result:
import matplotlib.pyplot as plt
plt.imshow(data.T, cmap=plt.cm.viridis)
plt.colorbar()
plt.show()
```
